```python
# import necessary libraries
import numpy as np
import pandas as pd
import plotly.express as px
import yfinance as yf
```

```python
# getting data from yahoo finance
df = yf.download(tickers='GLD', start='2020-01-01', end='2024-01-01')
df.tail()
```

```python
# plotting the close price of SPDR Gold Shares (GLD)
px.line(df, x=df.index, y='Close')
```

```python
# calculation of the log return
# pct_change gives us Net return: Rt=(Pt/Pt-1) - 1
# to calculate Log Return we have to use Gross Return: Rt + 1
df['Log-Ret'] = np.log(df['Close'].pct_change(1) + 1)
df.tail()
```

```python
# Simple Moving Average calculation
df['SMA-20'] = df['Close'].rolling(20).mean()
df.tail()
```

```python
# plotting the close price of SPDR Gold Shares (GLD) with SMA-20
px.line(df, x=df.index, y=['Close', 'SMA-20'])
```

```python
# Simple Moving Average calculation
df['SMA-60'] = df['Close'].rolling(60).mean()
df.tail()
```

```python
# plotting the close price of SPDR Gold Shares (GLD) with SMA-20 and SMA-60
px.line(df, x=df.index, y=['Close', 'SMA-20', 'SMA-60'])
```
